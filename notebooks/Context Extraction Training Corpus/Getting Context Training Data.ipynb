```python
graph = selventa_knowledge.get_graph()
```

```python
graph.summarize()
```

```python
grounded_graph = pybel.grounding.ground(graph)
grounded_graph.summarize()
```

```python
rows = []
for _, _, data in tqdm(grounded_graph.edges(data=True), leave=False):
    annotations = data.get(pc.ANNOTATIONS)
    evidence = data.get(pc.EVIDENCE)
    citation = data.get(pc.CITATION)
    if (
        not annotations
        or not evidence
        or not citation
        or citation[pc.CITATION_DB] != pc.CITATION_TYPE_PUBMED
    ):
        continue

    print(list(annotations))
        
    rows.append((
        citation[pc.CITATION_IDENTIFIER],
        evidence,
        list(annotations.get('Cell', [None]))[0],
        list(annotations.get('CellLine', [None]))[0],
        list(annotations.get('Anatomy', [None]))[0],
        list(annotations.get('Disease', [None]))[0],
    ))
    
df = pd.DataFrame(
    rows, 
    columns=[
        'pmid',
        'evidence',
        'cell',
        'cell_line',
        'anatomy',
        'disease',
    ],
).drop_duplicates()
df.head()
```
